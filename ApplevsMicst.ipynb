```python
data = {
    'Company': ['Apple', 'Microsoft'],
    'Total Net Sales': [85777, 75543],
    'Total Operating Expenses': [14326, 109433],
    'Net Income': [21448, 88136],
    'Total Assets': [331612, 118548],
    'Total Liabilities': [264904, 125286]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set the company name as the index for easy plotting
df.set_index('Company', inplace=True)

# Plotting financial metrics as a line plot (transpose to show trends per metric)
df.T.plot(kind='line', marker='o', figsize=(10, 6))

# Adding title and labels
plt.title('Company Financial Metrics Trend')
plt.xlabel('Financial Metrics')
plt.ylabel('Amount in Millions (USD)')
plt.xticks(rotation=0)

# Show the plot
plt.grid(True)
plt.show()
```
